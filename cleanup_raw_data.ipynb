```python
from datetime import datetime
from pyspark.sql.functions import col

catalog = "wikimedia_db"
uc_schema_raw_events = "raw_events"

# Keep today's suffix, example: "25_11_16"
today_suffix = datetime.now().strftime("%y_%m_%d")

# List all volumes in the schema
volumes_df = spark.sql(f"SHOW VOLUMES IN {catalog}.{uc_schema_raw_events}")

# Check what columns exist (debug):
print("Columns:", volumes_df.columns)

# Filter using the correct column name: volume_name
old_volumes = volumes_df.filter(~col("volume_name").endswith(today_suffix))

# Drop old volumes
for row in old_volumes.collect():
    vol_name = row["volume_name"]
    full_name = f"{catalog}.{uc_schema_raw_events}.{vol_name}"
    print(f"Dropping old volume: {full_name}")
    spark.sql(f"DROP VOLUME IF EXISTS {full_name}")
```
